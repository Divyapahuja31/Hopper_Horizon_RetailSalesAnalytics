## 1. Import Libraries
We start by importing the necessary libraries for data manipulation and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting aesthetic parameters for plotting
sns.set_theme(style="whitegrid")
%matplotlib inline
```

## 2. Load Dataset
Reading the raw CSV data from the `data/raw/` directory.

```python
file_path = '../data/raw/superstore.csv'
df = pd.read_csv(file_path)

print("Dataset loaded successfully!")
```

## 3. Basic Dataset Inspection
Performing a high-level check of the dataset's structure.

```python
# Shape of the dataset
print(f"Dataset Shape: {df.shape}")

# List of columns
print(f"\nColumns:\n{df.columns.tolist()}")

# Data types
print(f"\nData Types:\n{df.dtypes}")
```

## 4. Missing Values Check
Identifying if there are any null or missing values across the columns.

```python
missing_values = df.isnull().sum()
print("Missing Values per Column:")
print(missing_values[missing_values > 0] if missing_values.sum() > 0 else "No missing values found.")
```

## 5. Duplicate Rows Check
Checking for redundant data entries.

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

## 6. Summary Statistics
Generating descriptive statistics for numerical columns.

```python
df.describe().T
```

## 7. Unique Values for Key Categorical Columns
Exploring the categories available in the dataset.

```python
categorical_cols = ['Category', 'Sub-Category', 'Region', 'Segment']

for col in categorical_cols:
    print(f"\nUnique values in {col} ({df[col].nunique()}):")
    print(df[col].unique())
```
